```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True


def kth_prime(k):
    if not isinstance(k, int):
        raise TypeError("k має бути цілим числом")
    if k <= 0:
        raise ValueError("k має бути більше 0")

    count = 0
    num = 1

    while count < k:
        num += 1
        if is_prime(num):
            count += 1

    return num
```

```python
print(kth_prime(1))   # Очікувано: 2
print(kth_prime(5))   # Очікувано: 11
print(kth_prime(10))  # Очікувано: 29
```

```python
try:
    print(kth_prime("abc"))
except Exception as e:
    print("Помилка:", e)
```

```python
import matplotlib.pyplot as plt

ks = [1, 2, 3, 5, 10, 15]
values = [kth_prime(k) for k in ks]

plt.plot(ks, values, marker='o')
plt.title("Залежність k-го простого числа від k")
plt.xlabel("k")
plt.ylabel("k-те просте число")
plt.grid(True)
plt.show()
```
